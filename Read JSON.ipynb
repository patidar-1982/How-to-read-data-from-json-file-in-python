### Create JSON from dictionary

```python
# create dictionary
my_dict = {
    'Name':['Rahul','Anurag','Veer'],
    'ID':[1,2,3],
    'Marks_in_M1':[54,61,49],
    'Marks_in_M2':[49,55,48]
}
```

```python
#create data frome by dictionary
df4 = pd.DataFrame(my_dict)
```

```python
df4
```

```python
# convert this dataframe in json file
df_j = df4.to_json()
df_j
```

```python
# convert this dataframe in json file using orient parameter with split value
df_j = df4.to_json(orient='split')
df_j
```

```python
# convert JSON into dataframe without orient = 'split' we will get error
df5 = pd.read_json(df_j)
df5
```

```python
# convert JSON into dataframe with orient = 'split' ther will be no error
df5 = pd.read_json(df_j,orient='split')
df5
```

```python
# convert this dataframe in json file
df_j1 = df4.to_json()
df_j1
```

```python
df6 = pd.read_json(df_j1)
df6
```

```python
# convert this dataframe in json file using orient parameter with index value
df_j2 = df4.to_json(orient='index')
df_j2
```

```python
# convert JSON into dataframe without orient = 'index' 
df7 = pd.read_json(df_j2)
df7
```

```python
# convert JSON into dataframe with orient = 'index' 
df7 = pd.read_json(df_j2, orient='index')
df7
```

```python
#convert this dataframe in json file using orient parameter with record value
df_j3 = df4.to_json(orient='records')
df_j3
```

```python
# convert JSON into dataframe without orient = 'records' 
df8 = pd.read_json(df_j3)
df8
```

```python
# convert JSON into dataframe with orient = 'index' 
df8 = pd.read_json(df_j3, orient='records')
df8
```

```python
#convert this dataframe in json file using orient parameter with columns value
df_j4 = df4.to_json(orient='columns')
df_j4
```

```python
# convert JSON into dataframe with orient = 'index' 
df9 = pd.read_json(df_j4, orient='columns')
df9
```

```python
#convert this dataframe in json file using orient parameter with values value
df_j5 = df4.to_json(orient='values')
df_j5
```

```python
# convert JSON into dataframe with orient = 'index' 
df10 = pd.read_json(df_j5, orient='values')
df10
```

### Pandas have a nice inbuilt function called json_normalize() to flatten the simple to moderately semi-structured nested JSON structures to flat tables.

```python
# Example1
df = pd.DataFrame({
    'student_id': [1, 2, 3],
    'student_name': ['Alice', 'Bob', 'Chris'],
    'student_info': [
        {'gender': 'F', 'age': 20},
        {'gender': 'M', 'age': 22},
        {'gender': 'M', 'age': 33}
    ]
})
```

```python
df
```

```python
df2 = pd.json_normalize(df['student_info'])
```

```python
df3 = pd.concat([df,df2],axis=1)
```

```python
df3.drop(columns='student_info',inplace=True)
```
